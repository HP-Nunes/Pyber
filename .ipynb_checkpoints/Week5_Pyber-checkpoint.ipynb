```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import sem, ttest_ind 
from scipy.stats import linregress
import seaborn as sns

from matplotlib.backends.backend_pdf import PdfPages
```

```python
city = "raw_data/city_data.csv"
ride = "raw_data/ride_data.csv"
city_df = pd.read_csv(city)
ride_df = pd.read_csv(ride)
print(city_df.columns)
print(ride_df.columns)
#print(city_df.head())
#print(ride_df.head())
pyber_df = pd.merge(city_df, ride_df, on="city", how="left")
pyber_df.head()
```

```python
# I. Build a [Bubble Plot] that showcases the relationship between four key variables:

pyber_df_cities=pyber_df.loc[:,["city","fare","ride_id","driver_count","type"]]

#Mean: # Average Fare ($) Per City
pyber_df_cities_mean=pyber_df_cities.groupby("city").mean()
M_fare_cities=pyber_df_cities_mean["fare"] # *Avg Fare/City variable
#Count: # Total Number of Rides Per City
pyber_df_cities_count=pyber_df_cities.groupby("city").count()
S_rides_cities=pyber_df_cities_count["ride_id"] # *Number of rides/City variable
#Unique values: # Total Number of Drivers Per City # City Type (Urban, Suburban, Rural)
pyber_df_cities_unique=pyber_df_cities.groupby("city")
U_pyber_df_type=pyber_df_cities_unique["type"].unique() # *Types/City variable

pyber_df_drivers_unique=pyber_df_cities.groupby("city") # Pulls the unique value for the number of drivers in each City
U_pyber_df_drivers = pyber_df_drivers_unique["driver_count"].unique()
U_pyber_df_drivers.head()

pyber_df_drivers_unique=pyber_df_cities.groupby("city") # Pulls the unique value for the number of drivers in each City
U_pyber_df_drivers = pyber_df_drivers_unique["driver_count"].unique()
U_pyber_df_drivers=U_pyber_df_drivers.str[0] # *Unique Value for Driver Count per City
U_pyber_df_drivers.head()
```

```python
# % of Total Fares by City Type

pyber_df_type = pyber_df.loc[:,["city","fare","ride_id","driver_count","type"]]

#Mean: # % Total Fares ($) Per Type
#pyber_df_type_mean=pyber_df_type.groupby("type").mean() #Not trying to derive the % of avg. fare, but TOTAL fares.
pyber_type_sum=pyber_df_type.groupby("type").sum()
fare_type=pyber_type_sum["fare"] # *Avg Fare/Type variable
fare_type_perc = fare_type / fare_type.sum()

#Count: # Total Number of Rides Per Type
pyber_type_count =pyber_df_type.groupby("type").count() #We're counting the number of rides per type, not summing them.
rides_type=pyber_type_count["ride_id"] # *Number of rides/City variable
rides_type_perc = rides_type / rides_type.sum()

#Unique values: # Total Number of Drivers Per Type # City Type (Urban, Suburban, Rural)
pyber_type_unique=pyber_df_type.groupby(["type"]).sum()["driver_count"]# *Driver Count per Type
drivers_type_perc=pyber_type_unique/pyber_type_unique.sum()
#U_pyber_df_type=pyber_df_type_unique["driver_count"].unique()


type_pie_table = pd.DataFrame({"Total_Fare":fare_type_perc,
                                    "Total_Rides":rides_type_perc,
                               "Total_Drivers": drivers_type_perc
                                     })
type_pie_table.head()
```

```python
####################################################PIE CHARTS#################################################################
```

```python
#Pie Chart Attributes
colors = ["#FFD700", "#87CEFA", "#F08080"] #source: https://www.rapidtables.com/web/color/
labels = ["Rural","Suburban","Urban"]
```

```python
plt_pie_fare=plt.pie(
    # using data total)arrests
    type_pie_table['Total_Fare'],
    # with the labels being officer names
    labels=labels,
    # with no shadows
    shadow=False,
    # with colors
    colors=colors,
    # with one slide exploded out
    explode=(0, 0, 0.15),
    # with the start angle at 90%
    startangle=90,
    # with the percent listed as a fraction
    autopct='%1.1f%%',
)

plt.axis('equal')
plt.title("Share of the Total Fare ($) per City Type")

# View the plot
plt.tight_layout()

plt.show()

# with PdfPages('multipage_pdf.pdf') as pdf:
#                                             pdf.savefig()
#                                             plt.close()

```
